```python
import requests
import pandas as pd
import json
```

```python
url = 'http://overpass-api.de/api/interpreter?data=[out:json];(way[%22building%22](51.5175,-0.105,51.525,-0.09);node(w);way[%22building:part%22=%22yes%22](51.5175,-0.105,51.525,-0.09);node(w);relation[%22building%22](51.5175,-0.105,51.525,-0.09);way(r);node(w););out;'
```

```python
res = requests.get(url)
```

```python
items = res.json()
```

```python
pubs = filter(lambda i: 'tags' in i and 'amenity' in i['tags'] and 'pub' == i['tags']['amenity'], items['elements'])
```

```python
nodes = filter(lambda i: i['type'] == 'node', items['elements'])
```

```python
sos = filter(lambda i: i['id'] in pubs[0]['nodes'], nodes)
```

```python
items = []
for p in pubs:
    bits = filter(lambda i: i['id'] in p['nodes'], nodes)
    coords = [[i['lat'], i['lon']] for i in bits]
    items.append(coords)
json.dumps(items)
```
